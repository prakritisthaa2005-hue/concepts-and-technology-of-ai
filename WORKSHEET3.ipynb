3.3(A) % of total fare revenue by class
1. Total fare of all passengers

```python
total_fare = df["Fare"].sum()
```

2. Fare totals for each class

```python
fare_class1 = df[df["Pclass"] == 1]["Fare"].sum()
fare_class2 = df[df["Pclass"] == 2]["Fare"].sum()
fare_class3 = df[df["Pclass"] == 3]["Fare"].sum()

fare_list = [fare_class1, fare_class2, fare_class3]
fare_list
```

3. Percent from each class

```python
percentage = [f / total_fare * 100 for f in fare_list]
percentage
```

b.1. Create age group column

```python
def age_group(age):
    if age < 18:
        return "child"
    elif age < 65:
        return "adult"
    else:
        return "senior"

df["age_group"] = df["Age"].apply(age_group)
print(df[['Name','Age','age_group']].head())
```

total passenger

```python
total_passengers = len(df)
total_passengers
```

3. Count per group

```python
group_counts = df["age_group"].value_counts()
group_counts
```

4. Percentage

```python
group_percent = (group_counts / total_passengers) * 100
group_percent
```
