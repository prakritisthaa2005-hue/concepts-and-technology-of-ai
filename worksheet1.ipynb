**PROBLEM 4 — Matrix Operations**

```python
A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])
```

A = np.array([[3,4],[7,8]])
B = np.array([[5,3],[2,1]])


1. Prove A · A⁻¹ = I

```python
np.dot(A, np.linalg.inv(A))
```

2. Show AB ≠ BA

```python
np.dot(A,B)
np.dot(B,A)
```

3. Prove (AB)ᵀ = Bᵀ Aᵀ

```python
lhs = np.transpose(np.dot(A,B))
rhs = np.dot(B.T, A.T)
lhs, rhs
```

Matrix form: AX = B


```python
A = np.array([[2,-3,1],
              [1,-1,2],
              [3,1,-1]])

B = np.array([-1,-3,9])
```

Solution using inverse

```python
X = np.linalg.inv(A).dot(B)
X
```

Better method (recommended): solve()

```python
np.linalg.solve(A,B)
```
